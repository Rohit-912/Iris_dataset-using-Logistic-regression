```python
corr=df[['PetalLengthCm','SepalLengthCm','PetalWidthCm','SepalWidthCm']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation: Petal Length vs Sepal Length")
plt.show()
```
